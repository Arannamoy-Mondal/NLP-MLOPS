````python
import mlflow
````

````python
mlflow.set_tracking_uri("http://127.0.0.1:5000")
````

````python
mlflow.set_experiment("Local host connection")
````

````python
mlflow.set_experiment("test-2")
with mlflow.start_run():
    mlflow.log_metric("test-1",1)
    mlflow.log_metric("test-2",2)
    mlflow.log_metric("test-3",3)
````
