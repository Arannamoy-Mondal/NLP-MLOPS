```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentence=[
    "the glass of milk",
    "the glass of juice",
    "the cup of tea",
    "I prefer tea to coffee",
    "qml will change the paradigm of ml"
]
```

```python
voc_size=10000
```

```python
one_hot_repr=[one_hot(words,voc_size)for words in sentence]
one_hot_repr
```

```python
# word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence  import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sentance_length=8
embedded_docs=pad_sequences(one_hot_repr,padding="pre",maxlen=sentance_length)
print(embedded_docs)
```

```python
### feature representation
dim=10
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sentance_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
