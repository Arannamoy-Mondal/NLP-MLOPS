```python
from langchain.chat_models import init_chat_model
```

```python
model=init_chat_model("ollama:llama3.1:8b")
```

```python
res=model.invoke("What is quantum computing?")
res.content
```

```python
from langchain.tools import tool
```

```python
@tool
def get_weather(city:str)->str:
    """Get the weather at a location"""
    return f"The weather is sunny in {city}"


model_with_tools=model.bind_tools([get_weather])
```

```python
res=model_with_tools.invoke("What is the weather like in Chicago?")
print(res)

for tool_call in res.tool_calls:
    print(f"Tool: {tool_call['name']}")
    print(f"Tool: {tool_call['args']}")
```

### Tool execution

```python
messages=[{"role":"user","content":"What's the weather in Boston?"}]
ai_msg=model_with_tools.invoke(messages)
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_result=get_weather.invoke(tool_call)
    messages.append(tool_result)

final_res=model_with_tools.invoke(messages)
print(final_res.text)
```
