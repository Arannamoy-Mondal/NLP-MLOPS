```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
import pickle
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Arannamoy/datasets/refs/heads/main/Churn_Modelling/Churn_Modelling.csv")
```

```python
df.head(5)
```

```python
df=df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head(5)
```

```python
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
```

```python
df.head(10)
```

```python
onehot_encoder_geo=OneHotEncoder(handle_unknown="ignore")
geo_encoded=onehot_encoder_geo.fit_transform(df[['Geography']]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))

```

```python
geo_encoded_df
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
df.head(10)
```

```python
X=df.drop('EstimatedSalary',axis=1)
y=df['EstimatedSalary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
with open("label_encoder_gender.pkl","wb") as fl:
    pickle.dump(label_encoder_gender,fl)

with open("onehot_encoder_geo.pkl","wb") as fl:
    pickle.dump(onehot_encoder_geo,fl)

with open("scaler.pkl","wb") as fl:
    pickle.dump(scaler,fl)
```

### ANN Regression Problem Statement

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input,layers   
```

```python
model=Sequential(
    [
    Input(shape=(X_train.shape[1],)),
    Dense(64,activation=tf.keras.activations.relu),
    Dense(32,activation=tf.keras.activations.relu),
    Dense(1,activation=tf.keras.activations.relu)
    ]
)
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.001),loss=tf.losses.MeanAbsoluteError(),metrics=[tf.metrics.Accuracy()])
model.summary()
```

```python
from tensorflow.keras import callbacks
import datetime

log_dir="logs/regression/fit"+datetime.datetime.now().strftime("%Y%m%d%H%M")
```

```python
early_stopping_callback=callbacks.EarlyStopping(monitor='val_loss',patience=50,restore_best_weights=True)
```

```python
tensorboard_callback=callbacks.TensorBoard(log_dir,histogram_freq=1,)
```

```python
history=model.fit(
    X_train,y_train,
    validation_data=(X_test,y_test),
    epochs=100,
    callbacks=[early_stopping_callback,tensorboard_callback]
)
```

```python
test_loss,test_mae=model.evaluate(X_test,y_test)
print(f'Test MAE: {test_mae}')
```
