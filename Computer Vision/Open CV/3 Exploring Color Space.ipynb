```python
import cv2
```

```python
bgr_image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
```

```python
# convert bgr to gray
gray_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2GRAY)
cv2.imwrite("./Images/BGR2GRAY_OUTPUT.png",gray_image)
```

```python
# convert bgr to hsv
hsv_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2HSV)
cv2.imwrite("./Images/BGR2HSV_OUTPUT.png",hsv_image)
```

```python
# convert to lab
lab_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2LAB)
cv2.imwrite("./Images/BGR2LAB_OUTPUT.png",lab_image)
```

```python
# convert to rgb
rgb_image=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
cv2.imwrite("./Images/BGR2RGB_OUTPUT.png",rgb_image)
```
