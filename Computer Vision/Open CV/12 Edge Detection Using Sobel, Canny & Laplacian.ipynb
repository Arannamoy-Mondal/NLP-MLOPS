```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

### sobel edge detector

```python
sobel_x=cv2.Sobel(image_rgb,cv2.CV_64F,1,0,ksize=3)
sobel_y=cv2.Sobel(image_rgb,cv2.CV_64F,0,1,ksize=3)
sobel_combined=cv2.magnitude(sobel_x,sobel_y)
```

### laplacian detector

```python
laplacian=cv2.Laplacian(image_rgb,cv2.CV_64F,ksize=3)
laplacian_abs=cv2.convertScaleAbs(image_rgb,cv2.CV_64F)
```

### Canny

```python
canny_edges=cv2.Canny(image_rgb,180,200)
titles=['Original','Sobel','laplacian_all','Laplacian_abs','Canny']
images=[image,image_rgb,sobel_combined,laplacian,laplacian_abs,canny_edges]

j=0
plt.figure(figsize=(12,12))
for i in images:
    plt.subplot(3,2,j+1)
    j+=1
    plt.imshow(i)

plt.tight_layout()
plt.show()
```
