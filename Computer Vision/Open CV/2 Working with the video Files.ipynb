```python
import cv2
```

```python
video1=cv2.VideoCapture("./Video/test1.webm")
```

```python
if video1 is None:
    print("Error ! No video available")
else:
    print("Success")
```

```python
# display the video
while True:
    ret,frame=video1.read()
    if not ret:
        break
    cv2.imshow('Video window',video1)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break


video1.release()
cv2.destroyAllWindows()
```
