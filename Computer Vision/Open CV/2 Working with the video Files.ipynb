```python
# write video
video1=cv2.VideoCapture("Videos/test1.webm")
fourcc=cv2.VideoWriter.fourcc(*"mp4v")
width=int(video1.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(video1.get(cv2.CAP_PROP_FRAME_HEIGHT))
out=cv2.VideoWriter("Videos/output1.mp4",fourcc,300,(width,height))

while True:
    ret,frame=video1.read()
    if not ret:
        break
    out.write(frame)
out.release()
video1.release()
cv2.destroyAllWindows()
```
