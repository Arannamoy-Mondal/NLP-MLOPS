```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
image_grey=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
equalize_image=cv2.equalizeHist(image_grey)
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(4,4))
clahe_equlaize_image=clahe.apply(image_grey)


plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
plt.imshow(image_grey,cmap="grey")
plt.title("image_grey")
plt.subplot(1,3,2)
plt.imshow(equalize_image,cmap="grey")
plt.title("equalize_image")
plt.subplot(1,3,3)
plt.imshow(clahe_equlaize_image,cmap="grey")
plt.title("clahe_equlaize_image")
plt.show()
```

```python
image_yuv=cv2.cvtColor(image,cv2.COLOR_BGR2YUV)
image_yuv[:,:,0]=clahe.apply(image_yuv[:,:,0])
image_yuv=cv2.cvtColor(image,cv2.COLOR_YUV2BGR)
image_yuv=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)



plt.imshow(image_yuv)
```
