```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
kernel_s=5
average_blur=cv2.blur(image_rgb,(kernel_s,kernel_s))
plt.imshow(average_blur)
```

```python
gausianBlur_image=cv2.GaussianBlur(image_rgb,(kernel_s,kernel_s),0,0)
plt.imshow(gausianBlur_image)
```

```python
plt.imshow(cv2.medianBlur(image_rgb,kernel_s))
```
