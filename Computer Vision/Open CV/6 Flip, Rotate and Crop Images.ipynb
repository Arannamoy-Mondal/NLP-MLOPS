```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.subplot(2,2,1)
plt.imshow(image)
plt.subplot(2,2,2)
plt.imshow(image_rgb)
plt.tight_layout()
plt.show()
```

### Flip

```python
flip_horizontal=cv2.flip(image_rgb,1)
flip_vertical=cv2.flip(image_rgb,0)
flip_both_axes=cv2.flip(image_rgb,-1)

images=[image_rgb,flip_horizontal,flip_vertical,flip_both_axes]

plt.figure(figsize=(12,8))
for i in range(len(images)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])

plt.tight_layout()
plt.show()
```

### Rotate

```python
rotate90=cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotate90)
```

```python
rotate180=cv2.rotate(image_rgb,cv2.ROTATE_180)
plt.imshow(rotate180)
```

### Cropping and image

```python
print(image_rgb.shape)
x_start,y_start=400,300
x_end,y_end=800,900
cropped_image=image_rgb[y_start:y_end,x_start:x_end] 
plt.imshow(cropped_image)
plt.show() 
```
