#### In essence, stemming aims to remove unnecessary affixes (prefixes and suffixes) from words to get to their core meaning.

#### regexstemmer is commonly used in text data processing when you need to standardize words for analysis.

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
words=['Hellos','His','Byes','Newyorks','Californias','San jose']
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word)) 
```

```python
stemming.stem("Congratulations")
```

```python
stemming.stem("sitting")
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer("ing$|s$|e$|able$",min=4)
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("Hellos")
```

#### Snowball stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer=SnowballStemmer("english")
```

```python
for word in words:
    print(word," ",snowball_stemmer.stem(word))
```

```python
stemming.stem("fairly"),snowball_stemmer.stem("fairly")
```

```python
snowball_stemmer.stem("goes")
```
