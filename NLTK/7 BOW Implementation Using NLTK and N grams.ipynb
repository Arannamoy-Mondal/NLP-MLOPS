```python
corpus="""
Hey team, the meeting has been rescheduled to 3:00 PM today. Please find the updated agenda attached.
"""
```

```python
import re
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=nltk.sent_tokenize(corpus,"english")
```

```python
cleanded_corpus=[]

for i in range(0,len(corpus)):
    review=re.sub('[^a-zA-z]',' ',corpus[i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    cleanded_corpus.append(review)
    print(review)
```

## Create BOW

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
```

```python
x=cv.fit_transform(cleanded_corpus).toarray()
```
