### Lemmatization is the process of reducing a word to its base or dictionary form, often called its lemma.
### This is done to normalize words, making them consistent regardless of their inflected (different forms) or tensed states.

```python
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
```

```python
wordnet_lemmatizer=WordNetLemmatizer()
```

```python
'''
POS -
Noun -n
verb -v
adjective -a
adverb -r
'''
wordnet_lemmatizer.lemmatize("fairly",pos="a")
```
