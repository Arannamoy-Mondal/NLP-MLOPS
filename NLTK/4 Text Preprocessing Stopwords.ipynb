```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
stop_words=stopwords.words("english")
```

```python
stemmer=PorterStemmer()
```

```python
corpus="""
NVIDIA is a prominent technology company renowned for its leadership in graphics processing units (GPUs), system on a chip (SoC) products, and software. Founded in 1993, NVIDIA has become a driving force in the development of high-performance computing, artificial intelligence, and autonomous vehicles. Its GPUs are widely used in gaming, professional visualization, data centers, and various other applications, enabling advanced graphics rendering, machine learning, and scientific simulations. NVIDIA's innovations have significantly impacted industries such as gaming, data science, and autonomous driving, solidifying its position as a leader in the technology landscape.
"""
```

```python
sentences=nltk.sent_tokenize(corpus)
```

#### Apply filter and stopwords

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    sentences[i]=' '.join(words) # converting all the list of words in sentences
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowballstemmer.stem(word) for word in words if word not in stopwords.words('english')]
    sentences[i]=' '.join(words) # converting all the list of words in sentences
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word,pos="n") for word in words if word not in stopwords.words('english')]
    sentences[i]=' '.join(words) # converting all the list of words in sentences
```
