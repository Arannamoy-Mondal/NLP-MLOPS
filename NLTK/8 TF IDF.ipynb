## TF = Term Frequency

$$TF=\frac{\text{word count in doc}}{\text{total words in doc}}$$




	​


## IDF = Inverse Document Frequency
$$IDF=\log (\frac{\text{Total Documents}}{\text{Number of Documents containing the word}})=\log (\frac{\text{Total Sentences}}{\text{Number of Sentences containing the word}})$$

$$TF-IDF=TF×IDF$$
