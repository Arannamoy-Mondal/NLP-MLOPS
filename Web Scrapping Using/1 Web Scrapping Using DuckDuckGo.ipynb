```python
from ddgs import DDGS
import re
import time
```

```python
query="BTC latest price"
```

```python
def remove_html(text):
    return re.sub(r'<.*?>','',text)

def filter_dangerous(text):
    dangerous_phrases=[
        "ignore previous instructions",
        "send api key",
        "reveal secret",
        "system prompt",
        "confidential",
        "password",
    ]
    text_lower = text.lower()
    for phrase in dangerous_phrases:
        if phrase in text_lower:
            return None 
    return text



def truncate_text(text, max_chars):
    return text[:max_chars]


cleaned_results = []



with DDGS() as ddgs:
    results = ddgs.text(query, max_results=5)

    for r in results:
        title = r.get("title", "")
        url = r.get("href", "")
        body = r.get("body", "")


        body = remove_html(body)
        body = filter_dangerous(body)

        if body is None:
            continue  

        body = truncate_text(body, 10)

        cleaned_results.append({
            "title": title,
            "url": url,
            "content": body
        })

        time.sleep(1) 
```

```python
news=[]
for i,r in enumerate(results,1):
    news.append(f"{i}. {r['body']}")
    print(r['body'])
```
