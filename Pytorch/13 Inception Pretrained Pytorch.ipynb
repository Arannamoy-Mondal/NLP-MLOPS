```python
import torch
import torchvision.transforms as transforms
import torchvision.models as models
from PIL import Image
import matplotlib.pyplot as plt
import json
```

```python
model=models.inception_v3(pretrained=True)

model.eval()
```

### Image Transformation

```python
def transform_image(image_path):
    transform=transforms.Compose([  transforms.Resize((299,299)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406],std=[0.229,0.224,0.225])
        ])
    image=Image.open(image_path).convert('RGB')
    return transform(image).unsqueeze(0)
```

### Model Inferencing & Loading Class Labels

```python
import requests
```

```python
def predict(image_tensor,model):
    with torch.no_grad():
        outputs=model(image_tensor)
    return torch.nn.functional.softmax(outputs,dim=1)[0]

labels=[line.strip() for line in requests.get("https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt").text.splitlines()]
```

### Inferencing & Visualizing Results

```python
def visualize(image_path,predictions,labels):
    image=Image.open(image_path)
    plt.show("off")
    top5_prob,top5_catid=torch.topk(predictions,5)
    top5_labels=[labels[i] for i in top5_catid]
    plt.imshow(image)
    plt.title("\n".join([f"{label}:{prob:.2f}" for label,prob in zip(top5_labels,top5_prob)]))
    plt.show()

image_path="/home/workstation-p/Downloads/Projects/NLP-MLOPS/Computer_Vision/Open_CV/Images/green_mamba1.jpeg"
image_tensor=transform_image(image_path)
predictions=predict(image_tensor,model)
visualize(image_path,predictions,labels)
```
