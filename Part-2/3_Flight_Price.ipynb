```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder 
```

```python
p_df=pd.read_excel("Dataset/002 flight-price.xlsx")
df=p_df.copy()
```

```python
df.head(5)
```

```python
encoder=OneHotEncoder()
```

```python
pd.DataFrame(encoder.fit_transform(df[['Destination','Airline','Source']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df.drop_duplicates(inplace=True)
df[df.duplicated()]
```

```python
df['Journey_Date']=df['Date_of_Journey'].str.split("/").str[0].astype(int)
df['Journey_Month']=df['Date_of_Journey'].str.split("/").str[1].astype(int)
df['Journey_Year']=df['Date_of_Journey'].str.split("/").str[2].astype(int)
df.head(5)
```

```python
df['Dept_hour']=df['Dep_Time'].str.split(":").str[0]
df['Dept_minute']=df['Dep_Time'].str.split(":").str[1]
df.head(5)
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(":").str[0].astype(int)
df['Arrival_Minute']=df['Arrival_Time'].str.split(":").str[1].str.split(" ").str[0]
df['Arrival_Date']=df['Arrival_Time'].str.split(":").str[1].str.split(" ").str[1]
df['Arrival_Month']=df['Arrival_Time'].str.split(":").str[1].str.split(" ").str[2]
```

```python
df.head(5)
```

```python
df['Duration Hour']=df['Duration'].str.split(" ").str[0]
df['Duration Minute']=df['Duration'].str.split(" ").str[1]
```

```python
df.head(5)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].replace({'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'non-stop':0,np.nan:1}).astype(int)
```
