```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
image_grayscale=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png",cv2.IMREAD_GRAYSCALE)
plt.imshow(image_grayscale)
plt.show()
```

```python
x=-10
horizontal_kernel=np.array([[-1,-1,-1],[x,x,x],[1,1,1]])
vertical_kernel=np.array([[-1,x,1],[-1,x,1],[-1,x,1]])
diagonal_kernel=np.array([[x,-1,-1],[1,x,-1],[1,1,x]])
horizontal_filtered_image=cv2.filter2D(image_grayscale,-1,horizontal_kernel)
vertical_filtered_image=cv2.filter2D(image,-1,vertical_kernel)
diagonal_filtered_image=cv2.filter2D(image,-1,diagonal_kernel)

plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(horizontal_filtered_image,cmap="grey")
plt.subplot(1,3,2)
plt.imshow(vertical_filtered_image,cmap="grey")
plt.subplot(1,3,3)
plt.imshow(diagonal_filtered_image,cmap="grey")
```
