```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
```

```python
plt.imshow(image)
plt.show()
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2HSV))
```
