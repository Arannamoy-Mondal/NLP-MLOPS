```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12,8))
```

```python
plt.imshow(image_rgb)
plt.tight_layout()
plt.show()
```

```python
height, width= 500,700
blue=(255,127,0)
canvas=np.full((height,width,3),blue,dtype=np.uint8)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cv2.line(canvas,(50,50),(650,450),(255,255,255),5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cv2.line(canvas,(50,50),(650,450),(255,255,255),1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.show()
```

### Draw a rectangle

```python
cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),2)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),-1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.show()
```

### Draw a circle

```python
cv2.circle(canvas,(100,100),300,(0,0,255),-1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
plt.show()
```

### Draw an Ellipse


```python
cv2.ellipse(canvas,(350,350),(100,50),45,0,360,(0,255,0),3)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```
