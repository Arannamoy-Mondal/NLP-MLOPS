```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
# image resizing
print(image_rgb.shape)
```

```python
new_width,new_height= 2400,2400
resized_image=cv2.resize(image_rgb,(new_width,new_height))
print(resized_image.shape)
plt.imshow(resized_image)
```

### Image Scaling

```python
scale_percentage=50
s_width=int(image_rgb.shape[1]*scale_percentage/100)
s_height=int(image_rgb.shape[0]*scale_percentage/100)
scaled_image=cv2.resize(image_rgb,(s_width,s_height),interpolation=cv2.INTER_LANCZOS4)
print(scaled_image.shape)
plt.imshow(scaled_image)
```

### Interpolation

```python
resize_width=600
resize_height=600
methods={
    'nearest':cv2.INTER_NEAREST,
    'linear':cv2.INTER_LINEAR,
    'cubic':cv2.INTER_CUBIC,
    'area':cv2.INTER_CUBIC,
    'lancoz4':cv2.INTER_LANCZOS4
}
resized_image={}
plt.figure(figsize=(12,8))
for i,(name,method) in enumerate(methods.items()):
    resized_image[name]=cv2.resize(image_rgb,(resize_width,resize_width),method)
    plt.subplot(2, 3, i + 1)
    plt.imshow(resized_image[name])
    


plt.tight_layout()
plt.show()   
```
