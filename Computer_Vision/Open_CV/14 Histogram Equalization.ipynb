```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image=cv2.imread("./Images/Connection between Scikit Image, Open CV, Pillow, TF,Torch, YOLO.png")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
image_grayscale=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_grayscale)
plt.show()
```

```python
hist,bin=np.histogram(image_grayscale.flatten(),256,[0,256])


# histogram equalization

equalized_image=cv2.equalizeHist(image_grayscale)
equalized_hist,bins=np.histogram(equalized_image.flatten(),256,[0,256])

# plot the result

plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.imshow(image_grayscale,cmap="grey")
plt.title("Original Image")


plt.subplot(2,2,1)
plt.imshow(equalized_image,cmap="grey")
plt.title("equalized_image")

plt.subplot(2,2,2)
plt.plot(equalized_hist,color="green")
plt.title("Original Image")

plt.tight_layout()
plt.show()
```
