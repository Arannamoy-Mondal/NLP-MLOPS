## Transpose

```python
x=torch.arange(1,11).reshape(2,5)
print(f"{x}\n{x.t()}")
```

```python
x=torch.arange(1,11).reshape(2,5)
print(x.is_contiguous())
print(x.t().is_contiguous())
```

```python
x=torch.arange(1,11).view(2,5)
print(x.is_contiguous())
print(x.t().is_contiguous())
print(x.contiguous().view(5,2).is_contiguous())
```
