##### Features
- Solve classification (SVC) and regression (SVR)


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,roc_auc_score,accuracy_score
```

```python
X,y=make_classification(n_samples=10000,n_features=10,n_classes=2,n_clusters_per_class=2)
```

```python
sns.scatterplot(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="linear")
```

```python
svc.fit(X_train,y_train)
```

```python
print(classification_report(y_test,svc.predict(X_test)))
```

```python
print(accuracy_score(y_test,svc.predict(X_test)))
```

```python
rbf=SVC(kernel="rbf")
rbf.fit(X_train,y_train)
print(classification_report(y_test,rbf.predict(X_test)))
```

```python
rbf=SVC(kernel="sigmoid")
rbf.fit(X_train,y_train)
print(classification_report(y_test,rbf.predict(X_test)))
```

```python
rbf=SVC(kernel="poly")
rbf.fit(X_train,y_train)
print(classification_report(y_test,rbf.predict(X_test)))
```
